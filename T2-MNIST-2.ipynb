```
import numpy as np #algebra linear
import pandas as pd #processamento de dados
import tensorflow as tf #backend que roda sob o Keras
from tensorflow import keras #Keras está sendo incorporado ao Tensorflow
```

```
from keras.layers import Flatten #'aplaina' os dados em uma dimensão(?)
from keras.layers import Dense #todos as entradas se conectam aos neurônios
```

```
from keras.datasets import mnist #keras possui alguns datasets já formatados
(x_train,y_train), (x_test,y_test) = mnist.load_data() #formato do dataset
```

```
x_train = x_train / 255 #normaliza entre 0 e 1 os níveis dos pixels
x_test = x_test / 255
```

```
model = keras.Sequential() #modelo de rede neural sequencial
model.add(Flatten(input_shape=(28,28))) #os caracteres estão em imagens de 28x28 pixels
model.add(Dense(128,'relu')) #ReLu preserva valores positivos
model.add(Dense(10,'softmax')) #Softmax (*Sigmoid) confere relação de probabilidade
model.summary()
```

```
model.compile(optimizer='Adam',
              loss = 'sparse_categorical_crossentropy',
              metrics = ['accuracy']
             )
```

```
model.fit(x_train,y_train,epochs=5) #ajuste dos dados ao modelo compilado
```

```
previsao = model.predict(x=x_test)
previsao_r = np.argmax(previsao,axis=1) #o que ocorre aqui?
```

```
aleatorio = int(np.random.random()*len(previsao_r))

print("Previsao", len(previsao), "x", len(previsao[0]))
print("Numero aleatorio", aleatorio)
print("Elemento aleatorio na saída da rede", previsao[aleatorio])
print("Elemento aleatorio após argmax", previsao_r[aleatorio])
print("elemento aleatorio no teste", y_test[aleatorio])
```

```
acertos = 0
for i in range(len(y_test)):
  if previsao_r[i] == y_test[i]:
    acertos +=1

print("Tamanho da amostra:", len(y_test))
print("Acertos:", acertos)
print("Proporção de acertos:", acertos/len(y_test))

score = model.evaluate(x_test,y_test,verbose=0)
print("Score:", score)
```
